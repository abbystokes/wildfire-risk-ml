```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.enums import Resampling

import os

# Example FireScar path
tif_path = "capas/incendio/FireScars/FireScar_CL-ML_ID10006_u610_19860104.tif"  

# Open and display the image
with rasterio.open(tif_path) as src:
    image = src.read(1) 
    plt.figure(figsize=(8, 6))
    plt.imshow(image, cmap='inferno')  
    plt.title("Example FireScar TIF")
    plt.colorbar(label="Pixel Value")
    plt.axis('off')
    plt.show()
```

```python

# Open and resample the raster
with rasterio.open(tif_path) as src:
    scale_factor = 1 / 3  # 30m to 90m is downscaling by a factor of 3

    new_width = int(src.width * scale_factor)
    new_height = int(src.height * scale_factor)

    # Resample data to target shape
    image_resampled = src.read(
        1,
        out_shape=(new_height, new_width),
        resampling=Resampling.nearest  # Use average for continuous data, use nearest for categorical
    )

    # Scale the transform
    new_transform = src.transform * src.transform.scale(
        (src.width / new_width),
        (src.height / new_height)
    )

    # Plot the resampled image
    plt.figure(figsize=(8, 6))
    plt.imshow(image_resampled, cmap='inferno')
    plt.title("Resampled FireScar TIF (90m resolution)")
    plt.axis('off')
    plt.show()
```
